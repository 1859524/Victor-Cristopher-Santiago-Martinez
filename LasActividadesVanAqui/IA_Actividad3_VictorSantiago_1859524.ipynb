Ejercicio 1.1.- Piensa en 7 comidas diferente que te gusten. Implementa un programa que pregunte por las 7 comidas, las almacene en una lista y que posteriermente imprima cada una de ellas utilizando uno de los ciclos comentados en clase. 



```python
#Ejercicio 1.1.
a=input("Ingrese una comida que te guste #1: ")
b=input("Ingrese otra comida que te guste #2: ")
c=input("Ingrese otra comida que te guste #3: ")
d=input("Ingrese otra comida que te guste #4: ")
e=input("Ingrese otra comida que te guste #5: ")
f=input("Ingrese otra comida que te guste #6: ")
g=input("Ingrese la ultima comida que te guste #7: ")
lista=[a,b,c,d,e,f,g]
i = 0;
while i < 7:
  print(lista[i])
  i +=1
```
